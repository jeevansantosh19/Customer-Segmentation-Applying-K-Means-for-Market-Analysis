```python
# Encoding the Text Values
import sklearn
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
le
```

```python
# Encoding the Gender (Male = 0, Female = 1)
customer_data['Gender'] = le.fit_transform(customer_data['Gender'])
customer_data['Gender']
```

```python
# Preprocessing the Numerical Columns
from sklearn.preprocessing import MinMaxScaler

mms = MinMaxScaler()
mms
```

```python
# Scaling the Numerical Columns
customer_data[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']] = mms.fit_transform(customer_data[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']])
customer_data[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']]                                                                          
```

```python
customer_data
```

```python
# Unsupervised Machine Learning (K Means)
from sklearn.cluster import KMeans
```

```python
# Dropping the 'CustomerID' column
dataset1 = customer_data.drop('CustomerID', axis = 1)
dataset1
```

```python
# Feature Selection
X = dataset1[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']]
X
```

```python
# Initializing K Means
k_means = KMeans(n_clusters = 5, random_state = 42)
k_means
```

```python
# Training the KMeans
k_means.fit(X)
```

```python
# Elbow Method
wcss = []
for i in range(2, 6):
    k_means = KMeans(n_clusters = i, random_state = 42)
    k_means.fit(X)
    wcss.append(k_means.inertia_)

wcss
```

```python
# Creating a Column named 'Cluster'
dataset1['Cluster'] = k_means.fit_predict(X)
dataset1
```

```python
# Importing the Required libraries for Visualization
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Visualizing the Elbow Method
plt.figure(figsize = (12,12))
plt.plot(range(2, 6), wcss, marker = '*')
plt.title("Customer Segmentation")
plt.xlabel("Number Of Clusters")
plt.ylabel("WCSS")
plt.show()
```

```python
# Visualizing the Scatter Plot for Clusters
plt.figure(figsize=(12,12))

# Get unique cluster labels
clusters = dataset1['Cluster'].unique()

# Define a color map manually (optional)
colors = plt.cm.rainbow(np.linspace(0, 1, len(clusters)))

# Plot each cluster separately
for i, cluster in enumerate(clusters):
    cluster_data = dataset1[dataset1['Cluster'] == cluster]
    plt.scatter(cluster_data['Annual Income (k$)'],
                cluster_data['Spending Score (1-100)'],
                color=colors[i],
                label=f'Cluster {cluster}')

plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1 - 100)')
plt.title("Customer Segmentation Using K Means")
plt.legend()
plt.show()
```

```python
# Evaluating the Metrics
from sklearn.metrics import silhouette_score

kmeans_ss = silhouette_score(X, k_means.labels_)
print("\nSilhouette Score:", kmeans_ss)
```

```python
# Importing the Required Libraries for Serialization
import pickle
```

```python
# Saving the Model
with open('model/kmeans_model.pkl', 'wb') as file:
    pickle.dump({'model': k_means, 'score': kmeans_ss}, file)
```
